```python
import imageio
import numpy as np
import pylab
from sklearn.cluster import KMeans

np.random.seed(42)

filename = '3prgS6DyqB.mp4'
vid = imageio.get_reader(filename,  'ffmpeg')
NUM_FRAMES = vid.get_length()
width, height, _ = vid.get_data(0).shape

def summarizeVideo(vid):
    nums = [5, 9, 19, 29]
    for num in nums:
        image = vid.get_data(num)
        fig = pylab.figure()
        fig.suptitle('Frame #{}'.format(num+1), fontsize=20)
        pylab.imshow(image)
    pylab.show()
    
summarizeVideo(vid)
```

### Feature Creation

We consider a feature vector `(frame, x, y, R, G, B)` for each pixel.

```python
def createFeatureVector(vid):
    features = np.ndarray((NUM_FRAMES * width * height, 6))
    
    current = 0
    for i, frame in enumerate(vid):
        for x in range(width):
            for y in range(height):
                r, g, b = frame[x, y, :]
                features[current, :] = [i, x, y, r, g, b]
                current += 1
    return features
 
features = createFeatureVector(vid)
print("Data:", features)
```

```python
def kmeans(k, data):
    model = KMeans(init='k-means++', n_clusters=k, n_init=10)
    model.fit(data)
    centroids = model.cluster_centers_
    labels = model.labels_
    return centroids, labels
    
centroids, labels = kmeans(2, features)
print("Centroids:\n", centroids)
print("Labels:\n", labels)
```

```python
def generateClusteredVideo(features, centroids, labels, filename):
    out = imageio.get_writer(filename, 'ffmpeg')
    
    frame = np.ndarray((width, height, 3), dtype=np.uint8)
    for i, vec in enumerate(features):
        x, y = vec[1:3].astype(np.uint8)
        assn = centroids[labels[i]]
        frame[x, y, :] = assn[3:7].astype(np.uint8)
        if (x, y) == (63, 63):
            out.append_data(frame)
            frame = np.ndarray((width, height, 3), dtype=np.uint8)
            
    out.close()
    print("Finished writing to {}.".format(filename))

generateClusteredVideo(features, centroids, labels, 'k_2.mp4')        
```

```python
for k in [4, 6]:
    print("k-means with k = {}".format(k))
    centroids, labels = kmeans(k, features)
    generateClusteredVideo(features, centroids, labels, 'k_{}.mp4'.format(k))
```
