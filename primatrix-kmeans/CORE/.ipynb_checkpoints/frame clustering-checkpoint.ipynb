```python
import copy
import imageio
import numpy as np
import pylab
from sklearn.cluster import KMeans

np.random.seed(42)

filename = '7ivUhKeLMO.mp4'
vid = imageio.get_reader(filename,  'ffmpeg')
NUM_FRAMES = vid.get_length()
width, height, _ = vid.get_data(0).shape

def summarizeVideo(vid):
    image = vid.get_data(5)
    fig = pylab.figure()
    pylab.imshow(image)
    pylab.axis('off')
    pylab.show()
    
summarizeVideo(vid)
```

### Feature Creation

We consider a feature vector `(x, y, R, G, B)` for each pixel.

```python
def createFeatureVector(frame):
    features = np.ndarray((width * height, 5))
    current = 0
    for x in range(width):
        for y in range(height):
            r, g, b = frame[x, y, :]
            features[current, :] = [x, y, r, g, b]
            current += 1
    return features
 
features = createFeatureVector(vid.get_data(5))
print("Data:", features)
print(features.shape)
```

```python
def kmeans(k, data):
    model = KMeans(init='k-means++', n_clusters=k, n_init=10)
    model.fit(data)
    centroids = model.cluster_centers_
    labels = model.labels_
    return centroids, labels
    
centroids, labels = kmeans(2, features)
print("Centroids:\n", centroids)
print("Labels:\n", labels)
```

```python
def generateClusteredImage(features, centroids, labels, show=True):
    frame = np.ndarray((width, height, 3), dtype=np.uint8)
    for i, vec in enumerate(features):
        x, y = vec[0:2].astype(np.uint8)
        assn = centroids[labels[i]]
        frame[x, y, :] = assn[2:5].astype(np.uint8)
    if show:
        fig = pylab.figure()
        pylab.imshow(frame)
        pylab.axis('off')
        pylab.show()
    else:
        return frame

generateClusteredImage(features, centroids, labels)        
```

```python
for k in [4, 6]:
    print("k-means with k = {}".format(k))
    centroids, labels = kmeans(k, features)
    generateClusteredImage(features, centroids, labels)
```

```python
def scale_features(features, pos_scale=1):
    scale_features = copy.deepcopy(features)
    scale_features[:, 0:2] *= pos_scale
    return scale_features
    
for pos_scale in [0, 0.25, 0.5, 1, 2, 4, 8]:
    print("k-means with k = 2, position scale = {}".format(pos_scale))
    centroids, labels = kmeans(2, scale_features(features, pos_scale=pos_scale))
    generateClusteredImage(features, centroids, labels)
```

### HSV

HSV is a color format that may better represent human recognition over RGB.

http://www.cse.msu.edu/~pramanik/research/papers/2002Papers/icip.hsv.pdf

```python
from matplotlib.colors import rgb_to_hsv, hsv_to_rgb

# Conversion after deep copy
def rgbToHsvFeatures(features):
    hsv_features = copy.deepcopy(features)
    hsv_features[:, 2:5] = rgb_to_hsv(hsv_features[:, 2:5] / 255)
    return hsv_features
    
for pos_scale in [0, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]:
    print("k-means with k = 2, HSV, position scale = {}".format(pos_scale))
    centroids, labels = kmeans(2, scale_features(rgbToHsvFeatures(features), pos_scale=pos_scale))
    centroids[:, 2:5] = hsv_to_rgb(centroids[:, 2:5]) * 255
    generateClusteredImage(features, centroids, labels)
```

We see that HSV offers better segmentation than RGB for this particular image.  We can try to identify the effect of larger `k` for this particular scale.

```python
# In place conversion
def hsvToRgbFeatures(features):
    features[:, 2:5] = hsv_to_rgb(features[:, 2:5]) * 255
    return features
    
for k in [2, 3, 4, 6, 8, 10]:
    print("k-means with k = {}, HSV, position scale = 0.01".format(k))
    centroids, labels = kmeans(k, scale_features(rgbToHsvFeatures(features), pos_scale=0.01))
    generateClusteredImage(features, hsvToRgbFeatures(centroids), labels)
```

We see that `k=2` already provides substantial segmentation, and that segmentation resolution is negligibly improved with greater `k`.

### Video Rendering

We take our optimal `k=2` and `pos_scale=0.01` parameters and run k-means on each frame in the video.  We then combine the frames into a video.

```python
out = imageio.get_writer('frame_cluster_video_leopard_out.mp4', 'ffmpeg')
for frame in vid:
    features = createFeatureVector(frame)
    centroids, labels = kmeans(2, scale_features(rgbToHsvFeatures(features), pos_scale=0.01))
    out.append_data(generateClusteredImage(features, hsvToRgbFeatures(centroids), labels, show=False))
out.close()
```

### Duiker Sample

We'll now try this pipeline again with a duiker, a small antelope.



```python
filename = '3prgS6DyqB.mp4'
vid = imageio.get_reader(filename,  'ffmpeg')
NUM_FRAMES = vid.get_length()
width, height, _ = vid.get_data(0).shape

summarizeVideo(vid)

for pos_scale in [0, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]:
    print("k-means with k = 2, HSV, position scale = {}".format(pos_scale))
    features = createFeatureVector(vid.get_data(5))
    centroids, labels = kmeans(3, scale_features(rgbToHsvFeatures(features), pos_scale=pos_scale))
    centroids[:, 2:5] = hsv_to_rgb(centroids[:, 2:5]) * 255
    generateClusteredImage(features, centroids, labels)

out = imageio.get_writer('frame_cluster_video_duiker_out.mp4', 'ffmpeg')
for frame in vid:
    features = createFeatureVector(frame)
    centroids, labels = kmeans(3, scale_features(rgbToHsvFeatures(features), pos_scale=0.01))
    out.append_data(generateClusteredImage(features, hsvToRgbFeatures(centroids), labels, show=False))
out.close()
```

## RGV vs. HSV Summary

```python
opt_params = {'leopard': {'filename': '7ivUhKeLMO.mp4', 'k': 2, 'pos_scale': {'hsv': 0.01, 'rgb': 4}},
              'duiker': {'filename': '3prgS6DyqB.mp4', 'k': 3, 'pos_scale': {'hsv': 0.01, 'rgb': 0.5}},
              'primate': {'filename': '4E3MoZ4ZvK.mp4', 'k': 2, 'pos_scale': {'hsv': 0.01, 'rgb': 1}},
              'cattle': {'filename': '3DY9swri2b.mp4', 'k': 3, 'pos_scale': {'hsv': 0.005, 'rgb': 2}}}

for animal in opt_params:
    params = opt_params[animal]
    
    vid = imageio.get_reader(params['filename'],  'ffmpeg')
    NUM_FRAMES = vid.get_length()
    width, height, _ = vid.get_data(0).shape
    
    features = createFeatureVector(vid.get_data(5))
    print("Original")
    summarizeVideo(vid)
    print("RGB")
    centroids, labels = kmeans(params['k'], scale_features(features, pos_scale=params['pos_scale']['rgb']))
    generateClusteredImage(features, centroids, labels)
    print("HSV")
    centroids, labels = kmeans(params['k'], scale_features(rgbToHsvFeatures(features), pos_scale=params['pos_scale']['hsv']))
    generateClusteredImage(features, hsvToRgbFeatures(centroids), labels)
    
```
